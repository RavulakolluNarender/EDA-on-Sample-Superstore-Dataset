```r
#Imorting Required Libraries 

library(tidyverse)
library(lubridate)
library("ggplot2")
library(dplyr)
```

```r
#Reading the file from Kaggle

df = read.csv("../input/samplesuperstore/SampleSuperstore.csv")
```

```r
head(df)
```

```r
#Using Group By
df_ship = df %>% group_by(Ship.Mode, Category) %>%
                   summarise(Sum_of_Quantity = sum(Quantity),
                             .groups = 'drop')
 
View(df_ship)
```

```r
#Plotting a Bar plot  

ggplot(df_ship, aes(fill=Category, y=Sum_of_Quantity, x=Ship.Mode)) + 
    geom_bar(position="dodge", stat="identity")
```

```r
#Using Group By

df_state_sales = df %>% group_by(State) %>%
                   summarise(Maximum_Sales = max(Sales),
                             .groups = 'drop')
 
View(df_state_sales)
```

```r
# Plotting Bar plot
ggplot(df_state_sales, aes(y=Maximum_Sales, x=State)) +
        geom_bar(stat = "identity") +
        coord_flip()
```

```r
df_state_profit = df %>% group_by(State) %>%
                   summarise(Maximum_Profit = max(Profit),
                             .groups = 'drop')
 
View(df_state_profit)
```

```r
ggplot(df_state_profit, aes(y=Maximum_Profit, x=State)) +
        geom_bar(stat = "identity") +
        coord_flip()
```

```r
df_category_quantity = df %>% group_by(Category,Sub.Category) %>%
                   summarise(Sum_of_Quantity = sum(Quantity),                       
                             .groups = 'drop')
 
View(df_category_quantity)
```

```r
ggplot(df_category_quantity, aes(fill=Sub.Category, y=Sum_of_Quantity, x=Category)) + 
    geom_bar(position="dodge", stat="identity")
```

```r
df_category_sales = df %>% group_by(Category,Sub.Category) %>%
                   summarise(Maximum_Sales = max(Sales),                       
                             .groups = 'drop')
 
View(df_category_sales)
```

```r
ggplot(df_category_sales, aes(fill=Sub.Category, y=Maximum_Sales, x=Category)) + 
    geom_bar(position="dodge", stat="identity")
```

```r
df_category_profit = df %>% group_by(Category,Sub.Category) %>%
                   summarise(Maximum_Profit = max(Profit),                       
                             .groups = 'drop')
 
View(df_category_profit)
```

```r
ggplot(df_category_profit, aes(fill=Sub.Category, y=Maximum_Profit, x=Category)) + 
    geom_bar(position="dodge", stat="identity")
```
